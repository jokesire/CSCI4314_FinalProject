```python
import numpy as np
import matplotlib.pyplot as plt
import sys
```

```python
def cgr_genome(genome):
    corner_dict = {'a': np.array([0,0]), 'c': np.array([0,1]),
                   'g': np.array([1,1]), 't': np.array([1,0])}
    genome = genome.lower()
    xpos = [0.5]
    ypos = [0.5]
    curr_pos = np.array([0.5,0.5])
    for i,gene in enumerate(genome):
        vert = corner_dict[gene]
        diff = vert-curr_pos
        diff = diff/2
        curr_pos = curr_pos + diff
        xpos.append(curr_pos[0])
        ypos.append(curr_pos[1])
        if i%100 == 0:
            out_string = 'Progress: {}/{} \r'.format(i+1,len(genome))
            sys.stdout.write(out_string)
            sys.stdout.flush()
    return(xpos,ypos)
    
```

```python
genome_file = 'genome_efull.txt'
genome = ''
with open(genome_file,'r') as f:
    for f in f.readlines():
        f = f.strip('\n')
        genome = genome+f
```

```python
xpos,ypos = cgr_genome(genome)
```

```python
plt.plot(xpos,ypos,'k.', markersize = 0.005)
plt.title('Chaos Game Representation of E-coli k12')
plt.show()
```
